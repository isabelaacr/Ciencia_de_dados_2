**Distância interquartil**

```python
def conta_outliers(df):
  outliers_count = {} #armazenar a contagem de outliers por atributo
  for column in df.columns:
  #calcular a distância IQR
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

  #definir os limites para outliers
    limite_inf= Q1 - 1.5 * IQR
    limite_sup = Q3 + 1.5 * IQR

  #contar outliers para a coluna (feature) em questão
    outliers = df[(df[column] < limite_inf) | (df[column] > limite_sup)]
    outliers_count[column] = len(outliers)

  return outliers_count

```

```python
d = dados.copy()
del d['diagnosis']
c = conta_outliers(d)
outliers_counts_ordenado = dict(sorted(c.items(), key=lambda item: item[1], reverse=True))
outliers_counts_ordenado
```

**Técnicas de amostragem**

```python
dados = pd.read_csv('breast_cancer_custom.csv')

proporcao = dados['diagnosis'].value_counts(normalize=True)
proporcao
```

**Amostra estratificada**

```python
classes = dados['diagnosis'].unique()
tamanho_amostra = 10
qtd_por_classe = round(tamanho_amostra/len(classes))
amostras_por_classe = []
for c in classes:
  indices_c = dados['diagnosis'] == c
  obs_c = dados[indices_c]
  proporcao_c = len(obs_c) / len(dados)
  qnt_c = round(proporcao_c * tamanho_amostra)
  amostra_c = obs_c.sample(qnt_c)
  amostras_por_classe.append(amostra_c)

amostra_extratificada = pd.concat(amostras_por_classe)
amostra_extratificada
```
